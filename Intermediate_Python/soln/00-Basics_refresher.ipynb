# Basics Refresher

This exercise centres  around the loading and plotting data from a file while using Numpy functions. You have been given two dimensional data within a csv file and are required to plot the average, maximum and minimum of the columns of that data. 

1. Read the file-name using the Numpy loadtxt function
2. Find the mean, max and minimum of each column (axis =0) of the data 
3. Plot the mean minimum and maximum of each column with respect to the row number

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = 'data/inflammation.csv' 
```

```python
data = np.loadtxt(filename)
```

```python
mean = np.mean(data,axis=0)
maxn = np.max(data,axis = 0)
minn = np.min(data,axis = 0)
```

```python

fig = plt.figure(figsize=(10.0, 3.0))

axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.set_ylabel('average')
axes1.plot(data.mean(axis=0))

axes2.set_ylabel('max')
axes2.plot(data.max(axis=0))

axes3.set_ylabel('min')
axes3.plot(data.min(axis=0))

fig.tight_layout()
plt.show()
```
